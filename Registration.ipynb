```python
from tkinter import *
from openpyxl import *
import import_ipynb
from Record import record
from Train_model import train

wb = load_workbook('C:\\Users\\Prashu Gupta\\Desktop\\project\\Final_dataset\\excel.xlsx') 
sheet = wb.active
fields = ('Full Name', 'Enroll No', 'Address', 'Mobile')

def excel(): 
    sheet.column_dimensions['A'].width = 30
    sheet.column_dimensions['B'].width = 10
    sheet.column_dimensions['C'].width = 10
    sheet.column_dimensions['D'].width = 20

    sheet.cell(row=1, column=1).value = "Name"
    sheet.cell(row=1, column=2).value = "Enroll No"
    sheet.cell(row=1, column=3).value = "Address"
    sheet.cell(row=1, column=4).value = "Mobile"


def makeform(root, fields):
   lab = Label(root, text="Registration", anchor="center",font='Helvetica 18 bold')
   lab.pack(fill=X,side=TOP,pady = 25)
   entries = {}
   rowT=Frame(root)
   rowT.pack(side=TOP)
   for field in fields:
      row = Frame(rowT)
      row.pack(side=TOP)
      lab = Label(row, width=30, text=field+": ",font='Helvetica 12', anchor='w')
      ent = Entry(row, width=30, font='Helvetica 12')
      row.pack(side = TOP, padx = 20 , pady = 5)
      lab.pack(side = LEFT)
      ent.pack(side = RIGHT, expand = YES)
      entries[field] = ent
   row2=Frame(root)
   row2.pack(side=TOP)
   tkvar = StringVar(root)
   choices = { 'Voice 1','Voice 2','Voice 3','Voice 4','Voice 5','Voice 6'}
   tkvar.set('Voice 1') # set the default option
   popupMenu = OptionMenu(row2, tkvar, *choices)
   popupMenu.pack(side=LEFT, padx = 15 , pady = 15)
   bt=Button(row2,width=15,text="Add Voice",command=lambda : record(entries['Enroll No'].get(),tkvar.get(),rowP,sheet))
   bt.pack(side=LEFT,padx=20,pady = 15)
   rowP = Frame(row2,highlightbackground="grey", highlightcolor="grey", highlightthickness=2, width=100,height=30)
   rowP.pack(side=LEFT)
   trainM=Button(row2,text="Train",width=10,font='Helvetica 13',command=lambda :train(entries['Enroll No'].get()))
   trainM.pack(side=LEFT,padx=20,pady = 15)
   rowL=Frame(root)
   rowL.pack(side=TOP)
   submitbt=Button(rowL,text="Submit",font='Helvetica 13',command=lambda :train())
   submitbt.grid(row=0,column=0)
   Cancel=Button(rowL,text="Cancel",font='Helvetica 13',command=root.destroy)
   Cancel.grid(row=0, column = 1)
    
def insert(entries): 
    if (entries['Full Name'].get() == "" and entries['Enroll No'].get() == "" and entries['Address'].get() == "" and
        entries['Mobile'].get() == ""): 
        print("empty input") 

    else: 
        current_row = sheet.max_row 
        current_column = sheet.max_column 
        sheet.cell(row=current_row + 1, column=1).value = entries['Full Name'].get() 
        sheet.cell(row=current_row + 1, column=2).value = int(entries['Enroll No'].get())
        sheet.cell(row=current_row + 1, column=3).value = entries['Address'].get() 
        sheet.cell(row=current_row + 1, column=4).value = int(entries['Mobile'].get())

        # save the file
        wb.save('C:\\Users\\Prashu Gupta\\Desktop\\project\\Final_dataset\\excel.xlsx') 
        entries['Full Name'].focus_set() 

root = Tk()
ents = makeform(root, fields)
root.geometry("600x400")
root.mainloop()
```
